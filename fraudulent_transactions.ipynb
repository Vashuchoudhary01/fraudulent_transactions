```python
# here i changed the name on the column here

df = df.drop(["nameOrig", "nameDest"],axis = 1) # data preprocessing
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels

label_encoder = preprocessing.LabelEncoder()

# here i used both the fit and transform function for the type column in the given datset and change directly at the particular columns
df["type"] = label_encoder.fit_transform(df["type"])
```

```python
# using the loc function we are split the data
X, y = df.loc[:,df.columns != "isfraud"], df["isFraud"]
```

```python
# here i import the sklearn ML library for the specific purpose here train, test, split

from sklearn.model_selection import train_test_split


# as it is mentioned in the instruction i have to train the model on the 70% of the validation and rest for the test purpose


X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 44)
```

```python
# import another ML library functionality

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
# this basically for the test and transform the X_test and train the data
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# import Gaussian Naive Bayes model

from sklearn.naive_bayes import GaussianNB

from sklearn import metrics  # import the scikit learn metrics module for accuracy calculation

# Create a Gaussian Classifier
gnb = GaussianNB()

# Train the model using the training sets

gnb.fit(X_train, y_train)

# predict the response for test dataset here

y_pred = gnb.predict(X_test)
print("Accuracy of the model:", metrics.accuracy_score(y_test,y_pred))
```

```python
# use another model to test the accuracy of the model
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Accuracy of the model:", metrics.accuracy_score(y_test,y_pred))
```
